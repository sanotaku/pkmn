```python
import warnings

import numpy as np
import pandas as pd
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib
from sklearn.preprocessing import StandardScaler


warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('smoke.csv')

agecat = df['agecat'].values
deaths = df['deaths'].values
population = df['population'].values
smoke = df['smoke'].values


agecat_scaled = StandardScaler().fit_transform(agecat.reshape(-1, 1)).flatten()

df
```

```python
sns.scatterplot(x=agecat, y=deaths/population, hue=smoke)
plt.xlabel('年齢区分')
plt.ylabel('死亡率')
plt.show()
```

```python
sns.scatterplot(x=agecat, y=np.log(deaths/population), hue=smoke)
plt.xlabel('年齢区分')
plt.ylabel('log(死亡率)')
plt.show()
```

対数変換グラフより、線形ではないのでモデルになんらかの非線形項を入れる必要があるかもしれない<br>
とりあえずそのままオフセット項入りポアソン回帰で死亡者数をモデリングしてみる

```python
model = pm.Model()

with model:
    b1 = pm.Normal('b_1', mu=0, sigma=10)
    b2 = pm.Normal('b_2', mu=0, sigma=10)
    b3 = pm.Normal('b_3', mu=0, sigma=10)

    theta = b1 + b2 * smoke + b3 * agecat_scaled
    mu = pm.math.exp(theta) * population  # <-オフセット項

    obs = pm.Poisson('obs', mu=mu, observed=deaths)

pm.model_to_graphviz(model)
```

統計モデルは安易に割り算をすると情報が落ちてしまう。<br>
死亡率を計算してモデリングせず、観察者と死亡者をオフセット項を使用しそのままモデリングする。

```python
with model:
    iter = pm.sample(6000, tune=3000, idata_kwargs={'log_likelihood': True})
    iter.extend(pm.sample_posterior_predictive(iter, random_seed=8931, ))
```

```python
az.plot_trace(iter)
plt.tight_layout()
plt.show()

az.summary(iter)
```

```python
az.plot_posterior(iter, var_names=['b_2', 'b_3'], ref_val=0)
plt.tight_layout()
plt.show()
```

```python
figure = plt.figure(figsize=(12, 10))

for k in range(10):
    ax = figure.add_subplot(5, 2, k+1)
    az.plot_dist(iter.posterior_predictive['obs'][:, :, k].values.flatten())
    ax.axvline(deaths[k], color='r', linestyle='dashed')
    ax.set_title(f'AgeCategory: {agecat[k]} / Smoke: {smoke[k]}')

plt.tight_layout()
plt.show()
```

各カテゴリ(年齢カテゴリ×喫煙有無)ごとの推定分布と実際のデータ(赤線)で差が見られる。<br>
交互作用項と非線形項も考慮してみたモデルも考えてみる

### 交互作用項と非線形項を考慮したモデル

```python
agecat_scaled_sq = agecat_scaled * agecat_scaled
smoke_age = smoke * agecat_scaled
```

```python
model_extend = pm.Model()

with model_extend:
    b1 = pm.Normal('b_intercept', mu=0, sigma=10)
    b2 = pm.Normal('b_smoke', mu=0, sigma=10)
    b3 = pm.Normal('b_agecat', mu=0, sigma=10)
    b4 = pm.Normal('b_agecat_sq', mu=0, sigma=10)
    b5 = pm.Normal('b_smoke_agecat', mu=0, sigma=10)

    theta = b1 + b2 * smoke + b3 * agecat_scaled + b4 * agecat_scaled_sq + b5 * smoke_age
    mu = pm.math.exp(theta) * population  # <-オフセット項

    obs = pm.Poisson('obs', mu=mu, observed=deaths)

pm.model_to_graphviz(model_extend)
```

```python
with model_extend:
    iter_extend = pm.sample(6000, tune=3000, idata_kwargs={'log_likelihood': True})
    iter_extend.extend(pm.sample_posterior_predictive(iter_extend, random_seed=8931, ))
```

```python
az.plot_trace(iter_extend)
plt.tight_layout()
plt.show()

az.summary(iter_extend)
```

```python
az.plot_posterior(iter_extend, var_names=['b_smoke', 'b_agecat', 'b_agecat_sq', 'b_smoke_agecat'], ref_val=0)
plt.tight_layout()
plt.show()
```

```python
figure = plt.figure(figsize=(12, 10))

for k in range(10):
    ax = figure.add_subplot(5, 2, k+1)
    az.plot_dist(iter_extend.posterior_predictive['obs'][:, :, k].values.flatten())
    ax.axvline(deaths[k], color='r', linestyle='dashed')
    ax.set_title(f'AgeCategory: {agecat[k]} / Smoke: {smoke[k]}')

plt.tight_layout()
plt.show()
```

非線形項と交互作用項を入れたほうが合ってそう？

```python
df_waic = az.compare({'非線形項/交互作用項有り': iter_extend, '非線形項/交互作用項無し': iter}, ic='waic', scale='deviance')
az.plot_compare(df_waic)
plt.tight_layout()
plt.show()
df_waic
```

WAICからも非線形項と交互作用項を入れたほうがモデルとして適切そうである
