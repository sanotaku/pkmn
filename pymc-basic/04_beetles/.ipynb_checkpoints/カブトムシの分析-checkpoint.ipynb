```python
import warnings

import numpy as np
import pandas as pd
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib
from sklearn.preprocessing import StandardScaler


warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('beetle.csv')

gas = df['gas'].values
beetles = df['beetles'].values
deaths = df['deaths'].values
gas_scaled = StandardScaler().fit_transform(gas.reshape(-1, 1)).flatten()
```

```python
plt.scatter(x=gas, y=deaths / beetles)
plt.title('二硫化炭素ガス濃度によるカブトムシの死亡率変化')
plt.xlabel('二硫化炭素ガス濃度')
plt.ylabel('死亡率')
plt.show()
```

```python
model = pm.Model()

with model:
    a = pm.Normal('a', mu=0, sigma=10)
    b = pm.Normal('b', mu=0, sigma=10)
    mu = a * gas_scaled + b

    theta = pm.math.invlogit(mu)
    obs = pm.Binomial('obs', p=theta, n=beetles, observed=deaths)

pm.model_to_graphviz(model)
```

```python
with model:
    iter = pm.sample(3000, tune=1000, idata_kwargs={'log_likelihood': True})
    iter.extend(pm.sample_posterior_predictive(iter, random_seed=8931))
```

```python
az.plot_trace(iter, compact=False)
plt.tight_layout()
plt.show()

az.summary(iter)
```

```python
trace_num = 30
x_scaled_new = np.linspace(-2, 2, 50)

for idx in range(trace_num):
    a = iter.posterior['a'].values.flatten()[:trace_num][idx]
    b = iter.posterior['b'].values.flatten()[:trace_num][idx]
    mu = a * x_scaled_new + b
    p_array =  1 / (1 + np.exp(-mu))

    plt.plot(x_scaled_new, p_array, c='g', alpha=0.2)

plt.scatter(x=gas_scaled, y=deaths / beetles)
plt.title('二硫化炭素ガス濃度によるカブトムシの死亡率変化')
plt.xlabel('二硫化炭素ガス濃度')
plt.ylabel('死亡率')
plt.show()
```

```python
fig = plt.figure(figsize=(12, 8))

for k in range(8):
    ax = fig.add_subplot(4, 2, k+1)

    az.plot_dist(iter.posterior_predictive['obs'][0, :, k])
    ax.axvline(deaths[k], c='r', linestyle='dashed')
    ax.set_title(f'ガス濃度: {gas[k]}')

plt.tight_layout()
plt.show()
```
