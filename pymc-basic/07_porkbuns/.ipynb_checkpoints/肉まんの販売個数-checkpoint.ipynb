```python
import warnings

import numpy as np
import pandas as pd
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib
from sklearn.preprocessing import StandardScaler


warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('data.csv')

x = df['temperature'].values     # 気温
y = df['num_sold'].values        # 売上数
store_id = df['store_id'].values # 店舗ID
num_stores = max(store_id) + 1

x_scaled = StandardScaler().fit_transform(x.reshape(-1, 1)).flatten()

df.head(20)
```

```python
num_sold = y.reshape(num_stores, -1)
```

```python
plt.figure(figsize=(8, 4))
sns.heatmap(num_sold, annot=True, cmap='jet')
plt.xlabel('日付ID')
plt.xlabel('店舗ID')
plt.title('肉まんの売上数')
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.scatterplot(x=x[:10], y=num_sold.sum(axis=0))
plt.title('気温と全店舗合計売上数の関係')
plt.xlabel('気温')
plt.ylabel('全店舗合計売上数の関係')
plt.show()
```

```python
model = pm.Model()


with model:
    # 後ほどベイズ決定に使用するため、準備しておく
    shared_x_scaled = pm.Data('x_scaled', x_scaled)
    shared_store_id = pm.Data('store_id', store_id)
    shared_y = pm.Data('y', y)
    
    a = pm.Normal('a', mu=0, sigma=10)
    b = pm.Normal('b', mu=0, sigma=10)

    s = pm.HalfCauchy('s', beta=5)
    r = pm.Normal('r', mu=0, sigma=s, shape=(num_stores, ))

    theta = a * x_scaled + r[store_id] + b
    mu = pm.math.exp(theta)

    obs = pm.Poisson('obs', mu=mu, observed=y)


pm.model_to_graphviz(model)
```

```python
with model:
    iter = pm.sample(3000, tune=3000, idata_kwargs={'log_likelihood': True})
    iter.extend(pm.sample_posterior_predictive(iter, random_seed=8931))
```

```python
az.plot_trace(iter)
plt.tight_layout()
plt.show()

az.summary(iter)
```

```python
az.plot_posterior(iter, var_names=['a', 'b'])
plt.tight_layout()
plt.show()
```

```python
figure = plt.figure(figsize=(12, 16))


for k in range(30):
    ax = figure.add_subplot(10, 3, k+1)
    az.plot_dist(iter.posterior_predictive['obs'][:, :, k])
    ax.axvline(y[k], color='red', linestyle='dashed')
    ax.set_title(f'Temperature: {y[k]} / StoreID: {store_id[k]}')

plt.tight_layout()
plt.show()
```

```python
az.plot_violin(iter.posterior['r'], grid=(1, num_stores), figsize=(12, 4))
plt.show()
```

### 予測(ベイズ決定)をやってみる
1. 気温が5度の時、ある店舗(ID=6)で肉まんをいくつ蒸し器に入れておくか考える

```python
X_new = [5]         # 気温  : ５度
store_id_new = [6]  # 店舗ID:  6

x_scaled_new = (X_new - x.mean()) / x.std()

y_dummy = np.zeros_like(X_new, dtype=int)
```

```python
with model:
    pm.set_data({
        'x_scaled': x_scaled_new,
        'store_id': store_id_new,
        'y': y_dummy
    })
    ppc = pm.sample_posterior_predictive(iter.posterior, var_names=['obs'])
```

```python
num_sales_pred = np.array(ppc.posterior_predictive['obs'])
num_sales_pred
```

```python
plt.figure(figsize=(8, 4))
az.plot_dist(num_sales_pred)
plt.xlabel('売上数')
plt.ylabel('確率')
plt.tight_layout()
plt.show()
```

2. 気温が5度、標準偏差1の時、ある店舗(ID=6)で肉まんをいくつ蒸し器に入れておくか考える
