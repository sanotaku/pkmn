## 社員の出張回数をポアソン分布とゼロ過剰ポアソン分布で分析する
https://www.udemy.com/course/pymc3-basic/learn/lecture/26554620#overview

```python
import warnings

import pandas as pd
import pymc as pm
import arviz as az
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib


warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('data.csv')
X = df['count'].values
X
```

```python
plt.figure(figsize=(5, 3))
sns.histplot(X, discrete=True, shrink=0.7)
plt.xlim(-1, 5)
plt.xlabel('出張回数')
plt.show()
```

```python
X.mean(), X.var()  # 平均 = 分散では無さそう　 -> ポアソン分布ではダメかも
```

## とりあえずはポアソン分布に当てはめてみる

```python
model = pm.Model()

with model:
    model.add_coord('Data', mutable=True, values=df.index)

    x = pm.ConstantData('Count', X, dims='Data')
    mu = pm.HalfNormal('μ', sigma=10)
    obs = pm.Poisson('obs', mu=mu, observed=x, dims='Data')

pm.model_to_graphviz(model)
```

```python
with model:
    iter = pm.sample(3000, tune=1000, idata_kwargs = {'log_likelihood': True})
    iter.extend(pm.sample_posterior_predictive(iter, random_seed=8931))
```

```python
az.plot_trace(iter)
plt.tight_layout()
plt.show()
```

```python
az.plot_posterior(iter, figsize=(4, 3))
plt.tight_layout()
plt.show()
```

```python
az.plot_ppc(iter, num_pp_samples=1000, random_seed=8931)
plt.tight_layout()
plt.show()
```

少しデータから外れているようにも見える？

```python
iter.to_netcdf('iter.nc')
```

## ゼロ過剰ポアソン分布を使って当てはめ、WAICで比較する

```python
model_zero = pm.Model()


with model_zero:
    x = pm.ConstantData('Count', X)

    lam = pm.HalfNormal('λ', sigma=10)
    psi = pm.Uniform('ψ', lower=0, upper=1)
    obs = pm.ZeroInflatedPoisson('obs', mu=lam, psi=psi, observed=x)

pm.model_to_graphviz(model_zero)
```

```python
with model_zero:
    iter_zero = pm.sample(3000, tune=1000, idata_kwargs = {'log_likelihood': True})
    iter_zero.extend(pm.sample_posterior_predictive(iter_zero, random_seed=8931))
```

```python
az.plot_trace(iter_zero)
plt.tight_layout()
plt.show()
```

```python
az.plot_posterior(iter_zero)
plt.tight_layout()
plt.show()
```

```python
az.plot_ppc(iter_zero, num_pp_samples=1000)
plt.tight_layout()
plt.show()
```

### WAIC(広く使える情報量規準)で評価する

```python
# az.waic(iter)
# az.waic(iter_zero)
df_waic = az.compare({'ポアソン': iter, 'ゼロ過剰ポアソン': iter_zero}, scale='deviance', ic='waic')
df_waic
```

```python
az.plot_compare(df_waic)
plt.tight_layout()
plt.show()
```
